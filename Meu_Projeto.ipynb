```python
#Criando o DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Criando coluna de custo
df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df["Lucro"]  = df["Valor Venda"] - df["Custo Total"] 
```

```python
df.head()
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot(kind='pie', autopct='%0.2f%%', figsize=(5,5), label='Lucro x Ano')
plt.xlabel("Ano")
plt.ylabel("Receita")
num = 2
plt.show()
```

```python
total_atendimento_ano=df.groupby([df['Data Venda'].dt.year, 'Marca']).size()
total_atendimento_ano = total_atendimento_ano.unstack(level=0).fillna(0)
total_atendimento_ano.plot(kind='bar', figsize=(20,5))

```

```python
df.groupby([df['Data Venda'].dt.year, 'Marca']).size().unstack(1).fillna(0).plot(kind='bar')
```

```python
total_atendimento_ano=df.groupby([df['Data Venda'].dt.year, df['Data Venda'].dt.year]).size()
total_atendimento_ano = total_atendimento_ano.unstack(level=0).fillna(0)
total_atendimento_ano.plot(kind='bar', figsize=(20,5))
```
